```python
import pandas as pd

data = {
    "text": [
        # REAL NEWS
        "Government announces new education policy",
        "Scientists discover water on Mars",
        "New railway line inaugurated today",
        "Indian economy shows steady growth",
        "Health ministry releases new guidelines",
        "ISRO successfully launches satellite",
        "Supreme Court announces new verdict",
        "New technology introduced in colleges",
        "Weather department issues cyclone warning",
        "Vaccination drive expanded nationwide",
        "New metro line opened for public",
        "Electric vehicles encouraged by government",
        "Banks introduce new digital services",
        "Research shows benefits of regular exercise",
        "University announces exam schedule",

        # FAKE NEWS
        "Aliens landed in Chennai yesterday",
        "Drinking salt water cures all diseases",
        "Man becomes invisible after eating mango",
        "Earth will stop rotating tomorrow",
        "Mobile phones can charge using moonlight",
        "One fruit can make you live for 200 years",
        "Secret tunnel found under every school",
        "Humans can grow wings by special exercise",
        "Government giving free gold to everyone",
        "ATM machines will double your money",
        "This message will hack your phone",
        "Free money if you share this post",
        "Magic pill cures all illnesses",
        "Drinking only juice replaces food",
        "People can fly after drinking this juice"
    ],
    "label": [
        "real","real","real","real","real",
        "real","real","real","real","real",
        "real","real","real","real","real",
        "fake","fake","fake","fake","fake",
        "fake","fake","fake","fake","fake",
        "fake","fake","fake","fake","fake"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
X = df["text"]    # News text
y = df["label"]   # real / fake
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```

```python
news = [
    "Government announces new digital scheme",
    "People can become invisible using this trick",
    "Scientists release new climate report",
    "Drink this water to live forever"
]

news_vec = vectorizer.transform(news)
predictions = model.predict(news_vec)

for n, p in zip(news, predictions):
    print(n, "→", p)
```
